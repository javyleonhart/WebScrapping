```python
from bs4 import BeautifulSoup
import requests
```

```python
results = requests.get('https://cuspide.com/100-mas-vendidos/')
content = results.text

soup = BeautifulSoup(content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
infol = soup.find_all('div', class_='col small-6 large-2 max-w-large')
```

```python
titulos = []


for i in infol:
    titulo = i.h3.get_text(strip=True)
    titulos.append(titulo)
```

```python
autores = []

for i in infol:
    l = i.find('p', class_='author-product-loop')
    if l.a == None:
            autor = 'Sin autor'
            autores.append(autor)
    else:
        for pp in l:
            autor = pp.get_text(strip=True)
            autores.append(autor)
    
```

```python
urls = []
   
for i in infol:
   u = i.h3.a['href']
   urls.append(u)
```

```python
precios = []

for i in infol:
    l = i.find('div', class_='product-price')
    for pp in l:
        precio = pp.get_text(strip=True)
        precio = precio.strip("$")
        precio = precio.replace(".", "")
        precio = precio.replace(",", ".")
        if precio == '':
           precio = 0
           precios.append(precio)
        else:
            precio = float(precio)
            precio = int(precio)
            precios.append(precio)
```

```python
preciouss = []

for i in infol:
   u = i.h3.a['href']
   res = requests.get(u)
   cont = res.text
   sopa = BeautifulSoup(cont, 'html.parser')

   ur = sopa.find('div', class_='product-info summary col-fit col entry-summary product-summary text-left form-minimal')
   ur2 = ur.find_all('span')

   for i in ur:
      if 'Temporalmente sin stock' in i:
         pu = 0
         preciouss.append(pu)
         
      
   else:
      for i, elem in enumerate(ur2):
         if 'U$s ' in elem:
            pu = ur2[i].text
            preciouss.append(pu)
            print(ur2[i].text)   
    
```

```python
for i,e in enumerate(preciouss):
    if type(e) == str:
        elem = e.strip('U$s ')
        preciouss[i] = elem
```

```python
for i,e in enumerate(preciouss):
    if type(e) == str:
        elem = e.replace(',', ".")
        preciouss[i] = elem
```

```python
rd = requests.get('https://dolarapi.com/v1/dolares/blue')
c = rd.json()
dolar = list(c.values())[4]

```

```python
precioblue = []
for i in precios:
    p = i/dolar
    precioblue.append(p)
```
